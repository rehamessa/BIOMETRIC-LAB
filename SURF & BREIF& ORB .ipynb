>## SURF
Speeded Up Robust Features
* speeded-up version of SIFT
* approximates LoG with Box Filter\
* Convolution with box filter calculated 
* parallel for different scales

* Create surf object : surf= cv2.xfeatures2d.SURF_create(threshold)
* keypoints : kp= surf.detect(img)
* Compute the descriptors with surf : des= surf.compute(img, kp)
* Draw only keypoints : cv2.drawKeypoints(img, kp, None, color)

># BRIEF
 (Binary Robust Independent Elementary Features)
* BRIEF is a faster method feature descriptor calculation and matching. It also provides high recognition
rate unless there is large in-plane rotation.
* One important point is that BRIEF is a feature descriptor, it doesnâ€™t provide any method to find the
features.
* So you will have to use any other feature detectors like SIFT, SURF etc

>#
* Create STAR detector : star = cv2.xfeatures2d.StarDetector_create()
* find the keypoints : kp = star.detect(img,None)
* Initiate BRIEF extractor : brief = cv2.xfeatures2d.BriefDescriptorExtractor_create()
* compute the descriptors with BRIEF : kp, des = brief.compute(img, kp)
* Draw only keypoints : cv2.drawKeypoints(img,kp,None)

```python
import cv2
import matplotlib.pyplot as plt
img = cv2.imread('Screenshotfilter1.png',0)
# Initiate STAR detector
star = cv2.xfeatures2d.StarDetector_create()
# find the keypoints with STAR
kp = star.detect(img,None)
# Initiate BRIEF extractor
brief =cv2.xfeatures2d.BriefDescriptorExtractor_create()
# compute the descriptors with BRIEF
kp, des = brief.compute(img, kp)
img2=cv2.drawKeypoints(img,kp,None)

plt.subplot (122),plt.imshow (img2)
plt.title ('brief')
plt.show
```

># ORB
 (Oriented FAST and Rotated BRIEF)
 * alternative to SIFT and SURF in computation cost
 * ORB is basically a fusion of FAST keypoint detector

* Create ORB object : orb= cv2. ORB_create(max_num_feature) #500
* Find keypoints using orb: kp= surf.detect(img)
* Compute the descriptors with surf : des= orb.compute(img, kp)
* Draw only keypoints : cv2.drawKeypoints(img, kp, None, color)

```python
import cv2
img = cv2.imread('Screenshotfilter1.png',cv2.IMREAD_GRAYSCALE)
# Initiate STAR detector
orb = cv2.ORB_create()
# find the keypoints with ORB
kp = orb.detect(img, None)
# compute the descriptors with ORB
des = orb.compute(img, kp)
# draw only keypoints location,not size and orientation
img2 =cv2.drawKeypoints(img,kp,None,color=(255,0,0))
plt.subplot (122),plt.imshow (img2)
plt.title ('ORB')
plt.show
```

```python
orb = cv2.ORB_create(100)
kp = orb.detect(img, None)
des = orb.compute(img, kp)
img2 = cv2.drawKeypoints(img,kp,None,color=(255,0,0),flags=cv2.DrawMatchesFlags_DRAW_RICH_KEYPOINTS)
plt.subplot (122),plt.imshow (img2)
plt.title ('ORB')
plt.show
```
