># Edge detection
* identify points in a digital image 
* points where the image brightness varies sharply are called the edges
* Edges:
  * are significant local changes of intensity in an image.
  * Edges typically occur on the boundary between two different regions in an image.

## EDGE DETECTION USING LAPLACIAN DERIVATIVES.

```python
import cv2
from matplotlib import pyplot as plt
img =cv2.imread('Screenshotfilter1.png')
laplacian = cv2.Laplacian( img ,cv2.CV_64F)
plt.subplot (121),plt.imshow (img,cmap ="gray")
plt.title ('Original')
plt.subplot( 122),plt.imshow (laplacian,cmap = 'gray')
plt.title ('Edge')
plt.show
17
```

## EDGE DETECTION USING SOBEL DERIVATIVES.

```python
sobelx = cv2.Sobel(img , cv2.CV_64F, 1, 0,ksize =5)
sobely = cv2.Sobel(img , cv2.CV_64F, 1, 0,ksize =5)
sobelImg = cv2.add( sobelx , sobely)
plt.subplot (121),plt.imshow (img,cmap ="gray")
plt.title ('Original')
plt.subplot( 122),plt.imshow (sobelImg,cmap = 'gray')
plt.title ('sobel')
plt.show
```

## EDGE DETECTION USING CANNY.

```python
canny_= cv2.Canny(img,100,200)
plt.subplot (121),plt.imshow (img,cmap ="gray")
plt.title ('Original')
plt.subplot( 122),plt.imshow (canny_,cmap = 'gray')
plt.title ('Canny')
plt.show
```
