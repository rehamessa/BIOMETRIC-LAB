># Harris Corner Detection
* Determine which windows produce very large variations x and y direction.
* window found, a score R is computed
* applying a threshold to this score

># function cv2.cornerHarris() attribute :
* img - Input grayscale and float32 type.
* blockSize -the size of neighborhood.
* ksize - Sobel derivative.
* k -parameter equation.

```python
import matplotlib.pyplot as plt 
import cv2
import numpy as np
filename = ('Screenshotfilter1.png')
img = cv2.imread(filename)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
np.float32(gray)
dst = cv2.cornerHarris(gray, 2, 3, 0.04)
#result is dilated for marking the corners, not important
dst = cv2.dilate(dst, None)
# Threshold for an optimal value, it may vary depending on the image.
img[dst>0.01*dst.max()]=[0, 0, 255]
plt.subplot (122),plt.imshow (img)
plt.title ('harries')
plt.show
```

># Shi-Tomasi Corner Detection
 ### a function cv2.goodFeaturesToTrack()attribute:
* Img -grayscale image.
* N-max corners.
* X-number of corners you want to find (0-1).
* Y- minimum Euclidean distance

```python
import numpy as np
import cv2
img = cv2.imread('Screenshotfilter1.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
corners = cv2.goodFeaturesToTrack(gray,25,0.01,10)
corners = np.int0(corners)
for i in corners:
 x,y = i.ravel()
 cv2.circle(img,(x,y),3,(255,0,0),-1)
plt.subplot (122),plt.imshow (img)
plt.title ('shi-tomsi')
plt.show
```
