>## BGR :
* which each pixel has a triplet of values representing the blue, green, and red components.
* each pixel is represented by a triplet of 8-bit values, such as [0, 0, 0] for black, [255, 0, 0] for blue, [0, 255, 0] for green, [0, 0, 255] for
red, and [255, 255, 255] for white.

```python
#CONVERT FROM BGR TO GRAYSCALE.
import cv2
originalImage = cv2.imread( 'photo_2022-04-18_16-58-33.jpg' )
grayImage = cv2.cvtColor(originalImage, cv2.COLOR_BGR2GRAY)
cv2.imshow("original", originalImage)
cv2.imshow("gray", grayImage)
cv2.waitKey()
cv2.destroyAllWindows()
```

>## Grayscale :
* which pixel in a grayscale image is represented by a single 8-bit value, ranging from 0 for black to 255 for white.

```python
#CONVERT FROM GRAY TO BINARY
import cv2
originalImage = cv2.imread( 'photo_2022-04-18_16-58-33.jpg' )
#gray
grayImage = cv2.cvtColor(originalImage, cv2.COLOR_BGR2GRAY)
#binary
(thresh, blackAndWhiteImage) = cv2.threshold(grayImage, 127, 255,
cv2.THRESH_BINARY)
cv2.imshow("gray", grayImage)
cv2.imshow("binary", blackAndWhiteImage)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
## BGR TO RGB.
import cv2
originalImage = cv2.imread( 'photo_2022-04-18_17-30-14.jpg' )
RGBImage = cv2.cvtColor(originalImage, cv2.COLOR_BGR2RGB)
cv2.imshow("original", originalImage)
cv2.imshow("rgb", RGBImage)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#SPLIT $ MERGE
import cv2
originalImage = cv2.imread(
'photo_2022-04-18_17-30-14.jpg' )
B, G, R = cv2.split(originalImage)
cv2.imshow("original", originalImage)
cv2.imshow("blue", B)
cv2.imshow("Green", G)
cv2.imshow("red", R)
m=cv2.merge((B, G, R))
cv2.imshow("merged", m)
cv2.waitKey()
cv2.destroyAllWindows()
```

>## HSV :
* uses a different triplet of channels.
* Hue is the color's tone.
* saturation is its intensity.
* value represents its brightness

```python
#BGR TO HSV
import cv2
originalImage = cv2.imread( 'photo_2022-04-18_17-30-14.jpg' )
hsvImage=cv2.cvtColor(originalImage, cv2.COLOR_BGR2HSV)
cv2.imshow("original", originalImage)
cv2.imshow("HSV", hsvImage)
H=hsvImage[:,:,0]
S=hsvImage[:,:,1]
V=hsvImage[:,:,2]
cv2.imshow("hue", H)
cv2.imshow("saturation", S)
cv2.imshow("value", V)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
originalImage = cv2.imread( 'photo_2022-04-18_17-30-14.jpg' )
hsvImage=cv2.cvtColor(originalImage, cv2.COLOR_BGR2HSV)
cv2.imshow("original", originalImage)
lower = np.array([0, 100, 100])
upper = np.array([10, 255, 255])
Mask = cv2.inRange(hsvImage, lower, upper)
shape = cv2.bitwise_and(originalImage, originalImage, mask= Mask)
cv2.imshow("MAsk", Mask)
cv2.imshow("orange_shape",shape)
cv2.waitKey()
cv2.destroyAllWindows()
```

>## RESIZING IMAGE

```python
import cv2
image = cv2.imread( 'photo_2022-04-18_17-30-14.jpg' )
half = cv2.resize(image, (0, 0), fx = 0.5, fy = 0.5)
bigger = cv2.resize(image, (1050, 1610))
stretch_near = cv2.resize(image, (780, 540),
interpolation = cv2.INTER_NEAREST)#resampling using pixel area relation.
cv2.imshow("image", image)
cv2.imshow("half", half)
cv2.imshow("big", bigger)
cv2.imshow("stretch", stretch_near)
cv2.waitKey()
cv2.destroyAllWindows()
```

># Inversion

```python
import cv2
from matplotlib import pyplot as plt
img = cv2.imread("photo_2022-04-18_17-30-14.jpg",0)
img2 = 255-img #255 meaning maximum light
img_not = cv2.bitwise_not(img)#flips pixel values
plt.subplot(1, 3, 1), plt.imshow(img, cmap='gray'), plt.title('image')
plt.subplot(1, 3, 2), plt.imshow(img2, cmap='gray'), plt.title('invert1')
plt.subplot(1, 3, 3), plt.imshow(img_not, cmap='gray'), plt.title('invert2')
plt.show()
```
